```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/ML/train.csv')
df_test = pd.read_csv('/content/drive/MyDrive/ML/test.csv')
```

```python
cl = df['class'].value_counts()
cl_test = df_test['class'].value_counts()
```

```python
fig, ax = plt.subplots()
ax.set_title('Amount of data per class - Training')
ax.bar(['neither', 'depression', 'suicide'], [cl['neither'], cl['depression'], cl['suicide']])
```

```python
cl = df_test['class'].value_counts()
fig, ax = plt.subplots()
ax.set_title('Amount of data per class - Testing')
ax.bar(['neither', 'depression', 'suicide'], [cl['neither'], cl['depression'], cl['suicide']])
```
